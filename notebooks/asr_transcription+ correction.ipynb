```python
from pathlib import Path

AUDIO_DIR = Path("/content/drive/MyDrive/Digital_Speech/math_audio")

OUTPUT_DIR = Path("/content/drive/MyDrive/Digital_Speech/transcripts_raw")

OUTPUT_DIR.mkdir(exist_ok=True)
```

```python
import os

audio_files = list(AUDIO_DIR.glob("*.wav"))

print(f"Found {len(audio_files)} audio files")
```

```python
for audio_file in audio_files:

    output_file = OUTPUT_DIR / f"{audio_file.stem}.txt"

    # Skip if transcript already exists
    if output_file.exists():
        print(f"Skipping: {audio_file.name}")
        continue

    print(f"\nProcessing: {audio_file.name}")

    result = model.transcribe(
        str(audio_file),
        language="ar",
        fp16=True
    )

    transcript = result["text"]

    with open(output_file, "w", encoding="utf-8") as f:
        f.write(transcript)

    print(f"Saved: {output_file.name}")
```

```python
from google import genai

API_KEY = "API_KEY"

client = genai.Client(api_key=API_KEY)
```

```python
from pathlib import Path
import time

RAW_DIR = Path("/content/drive/MyDrive/Digital_Speech/transcripts_raw")
CORRECTED_DIR = Path("/content/drive/MyDrive/Digital_Speech/transcripts_corrected")

CORRECTED_DIR.mkdir(parents=True, exist_ok=True)

txt_files = list(RAW_DIR.glob("*.txt"))

print(f"Found {len(txt_files)} transcript files")
```

```python
def correct_arabic_transcript(text):
    prompt = f"""
You are an Arabic educational transcript correction assistant.

Correct the following Arabic ASR transcript from an educational mathematics video.

Instructions:
- Convert pronunciation-based Arabic into clear Modern Standard Arabic.
- Correct spelling, grammar, and punctuation.
- Remove duplicated words and ASR artifacts.
- Preserve the original educational meaning.
- Do not summarize.
- Do not add new information.
- Keep mathematical terms clear and accurate.
- Return only the corrected Arabic transcript.

Transcript:
{text}
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text
```

```python
for txt_file in txt_files:

    output_file = CORRECTED_DIR / txt_file.name

    if output_file.exists():
        print(f"Skipping: {txt_file.name}")
        continue

    print(f"\nCorrecting: {txt_file.name}")

    with open(txt_file, "r", encoding="utf-8") as f:
        raw_text = f.read()

    corrected_text = correct_arabic_transcript(raw_text)

    with open(output_file, "w", encoding="utf-8") as f:
        f.write(corrected_text)

    print(f"Saved: {output_file.name}")

    time.sleep(5)
```
